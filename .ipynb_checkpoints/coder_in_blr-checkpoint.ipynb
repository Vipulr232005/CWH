```python
with open ("initial_data.txt",encoding='utf-8')as f:
    data=f.read()
```

```python
data
```

```python
chunks=data.split("\n\n") 
```

```python
chunks[1]
```

```python
def parse_chunk(chunk):
    chunk = chunk.strip()
    sep_chunk = chunk.split('\n')

    if len(sep_chunk) < 6:
        return None

    username = sep_chunk[0]
    no_of_posts = int(sep_chunk[1].split(" post")[0].replace(",","" ))
    no_of_followers = sep_chunk[2]
    no_of_following = sep_chunk[3]
    name = sep_chunk[4]
    type_of_page = sep_chunk[5]
    bio = "\n".join(sep_chunk[6:])

    return {
        "username": username,
        "no_of_posts": no_of_posts,
        "no_of_followers": no_of_followers,
        "no_of_following": no_of_following,
        "name": name,
        "type_of_page": type_of_page,
        "bio": bio
    }
```

```python
all_chunks = []

for chunk in chunks:
    print(f"the chunk is :{chunk}")
    parsed_chunk = parse_chunk(chunk)

    if parsed_chunk is not None:
        all_chunks.append(parsed_chunk)

print(all_chunks)
```
